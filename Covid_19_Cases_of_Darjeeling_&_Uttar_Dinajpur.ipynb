```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
print('Setup Complete')
```

```
data_darj=pd.read_csv('Covid19(Darjeeling).csv', index_col='Date', parse_dates=True)
data_darj.tail()
```

```
data_ud=pd.read_csv('Covid19(Uttar Dinajpur).csv', index_col='Date', parse_dates=True)
data_ud.tail()
```

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases registered in a Single day at Darjeeling")
plt.xlabel("Date")
plt.ylabel('Covid-19 Cases registered in a Single day')
plt.stem(data_darj.index, data_darj['No of registered cases'])
```

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases registered in a Single day at Uttar Dinajpur")
plt.xlabel("Date")
plt.ylabel('Covid-19 Cases registered in a Single day')
plt.stem(data_ud.index, data_ud['No of registered cases'])
```

```
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 recoveries in Darjeeling & Uttar Dinajpur')
plt.xlabel("Date")
plt.ylabel('Total Covid-19 Cases Recovered')
sns.lineplot(data=data_darj['Total Recoveries'], label='Darjeeling', color='navy')
sns.lineplot(data=data_ud['Total Recoveries'], label='Uttar Dinajpur', color='teal')
```

```
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 Deaths in Darjeeling & Uttar Dinajpur')
plt.xlabel("Date")
plt.ylabel('Total no of Covid-19 Deaths')
sns.lineplot(data=data_darj['Total Deaths'], label='Darjeeling', color='black')
sns.lineplot(data=data_ud['Total Deaths'], label='Uttar Dinajpur', color='gray')
```

```
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 Cases in Darjeeling & Uttar Dinajpur')
plt.xlabel("Date")
plt.ylabel('Total Covid-19 Cases')
sns.lineplot(data=data_darj['Total Cases'], label='Darjeeling', color='green')
sns.lineplot(data=data_ud['Total Cases'], label='Uttar Dinajpur', color='red')
```
