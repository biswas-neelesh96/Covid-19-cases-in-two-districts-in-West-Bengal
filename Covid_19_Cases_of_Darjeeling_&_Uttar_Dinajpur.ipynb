Everyone knows about the rising of Covid-19 cases in the world & and our country; and as of now, our country, India is in the third highest no of Covid-19 cases registered. Everyone knows how Covid-19 has been impacted in our stated but hardly anyone knows the same about their home districts. With the help of **Spreadsheets (Google Sheets)** and **Data Visualization (Python)** tools, I tried to make a detailed analysis about the Covid-19 cases in two districts of West Bengal:

1) *Uttar Dinajpur*: Where I was born & had grew up.

2) *Darjeeling:* Where I'm currently living now

I took the record of daily cases of both thhe districts from reliable sources like [www.covid19india.org](https://www.covid19india.org) & [wbhealth.gov.in](https://wbhealth.gov.in) and made two separate spreadsheets in the form of **.csv** files, one for each district. The results are concluded on basis of:

1) **No. of cases registered, recoveries & deaths** recorded in both the districts separately.

2) **Total no. of cases, Total no. of recoveries, Total no. of deaths & Active Cases** by comparing both the districts in a single graph respectively.

3) **Recovery Rate & Fatality Rate** by comparing both the districts in a single graph respectively.

```
#import python libraries
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
print('Setup Complete')
```

1.a) Creating a dataframe showing the cases of the last five days in Darjeeling

```
data_darj=pd.read_csv('Covid19(Darjeeling).csv', index_col='Date', parse_dates=True)
data_darj.tail()
```

1.b) Creating a dataframe showing the cases of the last five days in Uttar Dinajpur

```
data_ud=pd.read_csv('Covid19(Uttar Dinajpur).csv', index_col='Date', parse_dates=True)
data_ud.tail()
```

2.a) Covid-19 Cases registered in a Single day at Darjeeling

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases registered in a Single day at Darjeeling as of 31st July 2020")
plt.ylabel('Covid-19 Cases registered in a Single day')
plt.stem(data_darj.index, data_darj['No of registered cases'])
```

2.b) Covid-19 Cases registered in a Single day at Uttar Dinajpur

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases registered in a Single day at Uttar Dinajpur as of 31st July 2020")
plt.xlabel("Date")
plt.ylabel('Covid-19 Cases registered in a Single day')
plt.stem(data_ud.index, data_ud['No of registered cases'])
```

3.a) Covid-19 Cases recovered in a Single day at Darjeeling

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases recovered in a Single day at Darjeeling as of 31st July 2020")
plt.xlabel("Date")
plt.ylabel('Covid-19 Cases recovered in a Single day')
plt.stem(data_darj.index, data_darj['No. of recoveries'],'g')
```

3.b) Covid-19 Cases recovered in a Single day at Uttar Dinajpur

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Cases recovered in a Single day at Uttar Dinajpur as of 31st July 2020")
plt.xlabel("Date")
plt.ylabel('Covid-19 Cases recovered in a Single day')
plt.stem(data_ud.index, data_ud['No. of recoveries'],'g')
```

4.a) Covid-19 Deaths in a Single day at Darjeeling

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Deaths in a Single day at Uttar Dinajpur as of 31st July 2020")
plt.xlabel("Date")
plt.ylabel('Covid-19 Deaths in a Single day')
plt.stem(data_ud.index, data_ud['No. of Deaths'],'r')
```

4.b) Covid-19 Deaths in a Single day at Uttar Dinajpur

```
plt.figure(figsize=(16,6))
plt.title("Covid-19 Deaths in a Single day at Darjeeling as of 31st July 2020")
plt.xlabel("Date")
plt.ylabel('Covid-19 Deaths in a Single day')
plt.stem(data_darj.index, data_darj['No. of Deaths'],'r')
```

5) Total no of Covid-19 recoveries in both the districts (comparison)

```
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 recoveries in Darjeeling & Uttar Dinajpur as of 31st July 2020')
plt.xlabel("Date")
plt.ylabel('Total Covid-19 Cases Recovered')
sns.lineplot(data=data_darj['Total Recoveries'], label='Darjeeling', color='navy')
sns.lineplot(data=data_ud['Total Recoveries'], label='Uttar Dinajpur', color='teal')
```

6) Total no of Covid-19 deaths in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 Deaths in Darjeeling & Uttar Dinajpur as of 31st July 2020')
plt.xlabel("Date")
plt.ylabel('Total no of Covid-19 Deaths')
sns.lineplot(data=data_darj['Total Deaths'], label='Darjeeling', color='black')
sns.lineplot(data=data_ud['Total Deaths'], label='Uttar Dinajpur', color='gray')
```

7) Total no of Covid-19 cases in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Total no of Covid-19 Cases in Darjeeling & Uttar Dinajpur as of 31st July 2020')
plt.xlabel("Date")
plt.ylabel('Total Covid-19 Cases')
sns.lineplot(data=data_darj['Total Cases'], label='Darjeeling', color='green')
sns.lineplot(data=data_ud['Total Cases'], label='Uttar Dinajpur', color='red')
```

7) Active cases in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Active Cases in Darjeeling & Uttar Dinajpur as of 31st July 2020')
plt.xlabel("Date")
plt.ylabel('No of Active cases')
sns.lineplot(data=data_darj['No. of Active Cases'], label='Darjeeling', color='orange')
sns.lineplot(data=data_ud['No. of Active Cases'], label='Uttar Dinajpur', color='brown')
```

7) Covid-19 recovery rate in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Recovery Rate in Darjeeling & Uttar Dinajpur as of 31st July 2020')
plt.xlabel("Date")
plt.ylabel('Recovery Rate (%)')
sns.lineplot(data=data_darj['Recovery Rate (%)'], label='Darjeeling', color='navy')
sns.lineplot(data=data_ud['Recovery Rate (%)'], label='Uttar Dinajpur', color='teal')
```

7) Covid-19 fatality rate in both the districts (comparison)

```
plt.figure(figsize=(16,6))
plt.title('Fatality Rate in Darjeeling & Uttar Dinajpur as of 31st July 2020')
plt.xlabel("Date")
plt.ylabel('Fatality Rate (%)')
sns.lineplot(data=data_darj['Fatality Rate (%)'], label='Darjeeling', color='black')
sns.lineplot(data=data_ud['Fatality Rate (%)'], label='Uttar Dinajpur', color='gray')
```
